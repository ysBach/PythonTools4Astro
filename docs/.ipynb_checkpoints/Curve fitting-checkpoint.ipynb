# Curve Fitting

When we do curve fitting, one most widely used package is ``scipy``. For non-linear function fitting to the data, ``scipy.optimize.curve_fit`` is a virtual standard. For simple polynomial fitting, one can also use ``numpy.polyfit``.


## ``curve_fit``
This ``curve_fit``, however, has one default setting that many astronomers may not have expected: the ``absolute_sigma``. Shortly put, ``absolute_sigma=False`` means that you are regarding the errors as relative weighting, i.e., you don't want to regard the errors as absolute (true) values. This is useful when you know the relative reliability of the data points. For example, you have magnitude data without error-bar from day 1 to 10. Although there is no actual errorbar, you may know that only day 1, 2, and 3 are reliable, and others are not reliable, due to the sky condition. Then you give *relative weighting* such as weight of 10 to day 1, 2, and 3, and weight of 1 to all other days. Then you can *force* the minimum chi square statistic to be exactly 1 by using some normalization process.

In some cases, however, you know absolute magnitude error, such as 10.1 mag +- 0.1 mag. In such cases, your minimum chi square statistic may not equal to 1. 

> FYI, chi square statistic in natural sciences is defined as $ \chi^2 := \sum_{i=1}^{N} \frac{(O_i - E_i)^2}{\sigma_i^2} $, where $O$ and $E$ are observed and expected values at i-th bin, respectively, and $N$ is the total number of observation.

By default, ``curve_fit`` uses ``absolute_sigma=False``, which goes same for MATLAB. I currently am not sure why it is set as default, but it is best to at least know this issue. There is also a [stack overflow thread](https://stackoverflow.com/questions/31704940/using-the-absolute-sigma-parameter-in-scipy-optimize-curve-fit).


I will demonstrate the possible confusion if you are not aware of this option. 

### Code 
In the following code, I did

* Define some useful functions:
  * Define a linear function to be used for ``curve_fit``
  * ``cov2err`` will convert the covariance matrix to error-bars: square root of diagonal components. FYI,
$ C = \begin{pmatrix}
   \sigma^2_1 & \sigma_{12} & \sigma_{13} & \cdots \\
   \sigma_{12} & \sigma^2_2 & \sigma_{23} & \cdots\\
   \sigma_{13} & \sigma_{23} & \sigma^2_3 & \cdots \\
   \vdots & \vdots & \vdots & \ddots
\end{pmatrix} ~.$
  * Define plotting and printing functions.
* A fake data (``data``) is made. Two different error-bar sets are made to see slight differences.
* Do curve fit to two data-error sets, with two options (``absolute_sigma = True`` or ``False``).
* So in total ``2*2=4`` results are generated.

```python
import numpy as np
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt

#%%
def linear(x, a=1, b=0):
    return a*x + b

def cov2err(cov):
    var_a = cov[0,0]
    var_b = cov[1,1]
    return np.sqrt(var_a), np.sqrt(var_b)

def plot_linear(axes, x, a, b, da, db, **kwargs):
    ''' Plot some of the possible fitting lines
    Plot lines based on best-fit parameters and errors.
    '''
    nums = 20
    arr_a = np.linspace(a-da, a+da+da/nums, num=nums)
    arr_b = np.linspace(b-db, b+db+db/nums, num=nums)
    for val_a in arr_a:
        for val_b in arr_b:
            axes.plot(x, linear(x, a=val_a, b=val_b), **kwargs)
    return

def print_result(datanum, absolute_sigma, a, b, da, db):
    string1 = 'Data {:d}, absolute_sigma={}'
    string2 = '\ta = {:+.4f} +- {:.4f} ({:+.4f}, {:+.4f})'
    string3 = '\tb = {:+.4f} +- {:.4f} ({:+.4f}, {:+.4f})'
    print(string1.format(datanum, absolute_sigma))
    print(string2.format(a, da, a-da, a+da))
    print(string3.format(b, db, b-db, b+db))

x = np.arange(10)
data = np.ones(10) # const (1) data value
data[[0,1,2]] -= 0.1
data[[7,8,9]] += 0.1
err0 = data/10     # const (0.1) errorbar
err1 = err0.copy()
err1[[3, 5, 7]] += 0.1

(a0_F, b0_F), pcov0_F = curve_fit(linear, x, data, sigma=err0, absolute_sigma=False)
(a0_T, b0_T), pcov0_T = curve_fit(linear, x, data, sigma=err0, absolute_sigma=True)
(a1_F, b1_F), pcov1_F = curve_fit(linear, x, data, sigma=err1, absolute_sigma=False)
(a1_T, b1_T), pcov1_T = curve_fit(linear, x, data, sigma=err1, absolute_sigma=True)

da0_F, db0_F = cov2err(pcov0_F)
da0_T, db0_T = cov2err(pcov0_T)
da1_F, db1_F = cov2err(pcov1_F)
da1_T, db1_T = cov2err(pcov1_T)

print_result(0, False, a0_F, b0_F, da0_F, db0_F)
print_result(0, True , a0_T, b0_T, da0_T, db0_T)
print_result(1, False, a1_F, b1_F, da1_F, db1_F)
print_result(1, True , a1_T, b1_T, da1_T, db1_T)

fig, axs = plt.subplots(2,1, sharey=True, sharex=True, figsize=(8,8))
axs[0].errorbar(x, data, yerr=err0, capsize=3, marker='x', ls=':')
axs[1].errorbar(x, data, yerr=err1, capsize=3, marker='x', ls=':')
axs[0].set_title("Data 0")
axs[1].set_title("Data 1")
axs[0].text(0, 1.25, "absolute_sigma=False", color='r', fontsize=16)
axs[0].text(0, 1.15, "absolute_sigma=True", color='b', fontsize=16)

plot_linear(axs[0], x, a0_F, b0_F, da0_F, db0_F, color='r', alpha=0.1)
plot_linear(axs[0], x, a0_T, b0_T, da0_T, db0_T, color='b', alpha=0.1)
plot_linear(axs[1], x, a1_F, b1_F, da1_F, db1_F, color='r', alpha=0.1)
plot_linear(axs[1], x, a1_T, b1_T, da1_T, db1_T, color='b', alpha=0.1)

plt.show()
```

### Discussion

First of all, note that the **best-fit parameters do NOT change with respect to the ``absolute_sigma`` option**, as expected. But **the uncertainties in the fitted parameters DO change**. The uncertainty for ``absolute_sigma=False`` (default, call it **case F**; opponent as **case T**) case will never change unless we change ``err0`` or ``err1`` so that the relative uncertainty changes. 

For example,
* ``err0 += 0.1`` will change the uncertainty for *both* case T and F.
* ``err0 *= 0.1`` will change the uncertainty for *only* case T.

Now you can easily see that changing ``absolute_sigma`` from ``False`` (case F; default option) to ``True`` will result in larger uncertainties in both slope and intercept (``a`` and ``b``, respectively). Check the image (blue for ``True`` option and red for ``False`` option) and the printed result. 

If the array ``err0`` is decreased by factor of 100, for example, the resulting errors in slope and intercept decreases significantly for the case T, and the uncertainties are smaller than the case F. You may expect the fitted line should be more accurate if error gets smaller and vice versa; then you **must explicitly set ``absolute_sigma=True``**).
